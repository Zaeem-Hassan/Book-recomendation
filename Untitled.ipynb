```python
rating =  pd.read_csv("Ratings.csv")
rating.rename(columns =  {'User-ID':'id','Book-Rating':'rating'},inplace = True)
```

```python
rating.head()
```

```python
x =  rating['id'].value_counts() > 200
```

```python
x = x[x ==  True]
```

```python
y = x.index
```

```python
rating = rating[rating['id'].isin(y) == True]
```

```python
book_rating =  rating.merge(book, on='ISBN' )
```

```python
book_rating.shape
```

```python
book_rating.sample(5)
```

```python
rating_number = book_rating.groupby('title')['rating'].count().reset_index()
```

```python
rating_number.rename(columns = {'rating':'number of rating'},inplace = True)
```

```python
rating_number
```

```python
df = rating_number.merge(book_rating,on = 'title')
```

```python
df.head()
```

```python
df.shape
```

```python
df = df[df['number of rating']>=50]
```

```python
df.shape
```

```python
df.drop_duplicates(['title','id'],inplace = True)
```

```python
df.shape
```

```python
df.tail()
```

```python
pivot =  df.pivot_table(columns ='id',index= 'title',values='rating')
```

```python
pivot.head()
```

```python
pivot.fillna(0,inplace= True)
```

```python
pivot.head()
```

```python
sparse  = csr_matrix(pivot)
type(sparse)
```

```python
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(sparse)
```

```python
distance, suggestion = model.kneighbors(pivot.iloc[54,:].values.reshape(1,-1),n_neighbors = 6)
```

```python
def recommend_book(book_name):
    try:
        book_id = np.where(pivot.index == book_name)[0][0]
        distance, suggestions = model.kneighbors(pivot.iloc[book_id, :].values.reshape(1, -1), n_neighbors=6)
        print(f"Recommendations for '{book_name}':")
        for i in range(1, len(suggestions.flatten())):
            print(f"- {pivot.index[suggestions.flatten()[i]]}")

    except IndexError:
        print(f"Book '{book_name}' not found in the database.")
```

```python
recommend_book('Animal Farm')
```
